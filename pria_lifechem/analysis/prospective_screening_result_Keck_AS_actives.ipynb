```python
import pandas as pd
import numpy as np
from IPython.display import HTML
from collections import OrderedDict
```

```python
dir_ = '../../output/stage_2_predictions/Keck_Pria_AS_Retest'
complete_df = pd.read_csv('{}/complete_prediction.csv'.format(dir_))
rank_df = pd.read_csv('{}/complete_rank.csv'.format(dir_))
```

```python
test_data_df = pd.read_csv('../../dataset/keck_lc4.csv.gz')
test_data_df = test_data_df[['Molecule', 'SMILES']]
rank_df = rank_df.merge(test_data_df, how='left', left_on='molecule id', right_on='Molecule', sort=False)
```

```python
model_names = rank_df.columns[3:-2]
```

```python
dataframe = pd.read_excel('../../output/stage_2_predictions/Keck_LC4_export.xlsx')

molecule_name_list = dataframe['Molecule Name'].tolist()
supplier_id = dataframe['Supplier ID'].tolist()
failed_id = ['F0401-0050', 'F2964-1411', 'F2964-1523']
inhibits = dataframe[
    'PriA-SSB AS, normalized for plate and edge effects, correct plate map: % inhibition Alpha, normalized (%)'].tolist()
neo_dataframe = pd.read_csv('../../output/stage_2_predictions/pria_lc4_retest_may18.csv')
failed_molecule_names = neo_dataframe[neo_dataframe['Active'] == 0]['Row Labels'].tolist()
failed_molecule_names += ['SMSSF-0044356', 'SMSSF-0030688']

positive_enumerate = filter(lambda x: x[1] >= 35 and supplier_id[x[0]] not in failed_id and molecule_name_list[x[0]] not in failed_molecule_names, enumerate(inhibits))
positive_idx = map(lambda x: x[0], positive_enumerate)
```

# Show n-hits in top 250

```python
hits_id_dict = OrderedDict()
hits_molecule_dict = OrderedDict()
hits_smiles_dict = OrderedDict()

hits_id_dict['All'] = positive_idx
order = rank_df[['molecule id', 'SMILES']].loc[positive_idx]
hits_molecule_dict['All'] = order['molecule id'].loc[positive_idx].tolist() 
hits_smiles_dict['All'] = order['SMILES'].loc[positive_idx].tolist()

for (idx, model_name) in enumerate(model_names):
    order = rank_df[['molecule id', 'SMILES', model_name]].loc[positive_idx]
    hits_id_tuple = filter(lambda x: 0<=x[1] and x[1] < 250, order[model_name].iteritems())
    hits_id = [x[0] for x in hits_id_tuple]
    hits_id_dict[model_name] = hits_id
    print model_name, 'hits'
    print hits_id
    print
    hits_molecule_dict[model_name] = order['molecule id'].loc[hits_id].tolist()
    hits_smiles_dict[model_name] = order['SMILES'].loc[hits_id].tolist()
```

```python
for model_name in hits_molecule_dict.keys():
    print model_name,':\t',  len(hits_molecule_dict[model_name])
    print hits_molecule_dict[model_name]
    print
```

```python
for model_name in hits_smiles_dict.keys():
    temp_df = pd.DataFrame()
    temp_df['molecule id'] = hits_molecule_dict[model_name]
    temp_df['SMILES'] = hits_smiles_dict[model_name]
    temp_df.to_csv('./plottings/prospective_screening_model_hits/actives_in_{}.csv'.format(model_name), index=None)
```
