# Plot EF Curve

```python
from EF_curve_support import *

%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20.0, 30.0)
```

```python
json_file_list = ['single_classification_22.json',
                  'single_classification_42.json',
                  'single_regression_2.json',
                  'single_regression_11.json',
                  'multi_classification_15.json',
                  'multi_classification_18.json']

json_paths = ['../../output/cross_validation_RMI/{}'.format(p) for p in json_file_list]
print json_paths

weights_dir_paths = ['../../output/cross_validation_RMI/single_classification_22/46244865/',
                     '../../output/cross_validation_RMI/single_classification_42/46101892/',
                     '../../output/cross_validation_RMI/single_regression_2/46245031/',
                     '../../output/cross_validation_RMI/single_regression_11/46245032/',
                     '../../output/cross_validation_RMI/multi_classification_15/46245054/',
                     '../../output/cross_validation_RMI/multi_classification_18/46245048/']

model_names_list = ['single_classification_22',
                    'single_classification_42',
                    'single_regression_2',
                    'single_regression_11',
                    'multi_classification_15',
                    'multi_classification_18']
```

```python
EF_ratio_list = np.linspace(0.001, 0.15, 100)
N = len(json_file_list)
data_set_name = 'Keck_RMI'

data_pd_list = []
for i in range(N):
    temp_pd = get_EF_curve_in_pd(EF_ratio_list=EF_ratio_list,
                                 data_set_name=data_set_name,
                                 config_json_file=json_paths[i],
                                 weight_dir=weights_dir_paths[i],
                                 model_name=model_names_list[i],
                                 regenerate=False)
    data_pd_list.append(temp_pd)
    
whole_pd = data_pd_list[0]
for i in range(1, N):
    whole_pd = whole_pd.append(data_pd_list[i])
```

```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20.0, 15.0)

# Calculate Normalized EF, and add it to dataframe
whole_pd['Normalized EF'] = whole_pd.apply(lambda row: row['EF'] / row['EF max'], axis=1)
sns.tsplot(data=whole_pd, time="EFR", unit="running process", condition="model", value="Normalized EF")
plt.savefig('plottings/NEF_curve.png')
```

```python
focused_pd = whole_pd[whole_pd['EFR']<=0.06]
sns.tsplot(data=focused_pd, time="EFR", unit="running process", condition="model", value="Normalized EF")
```
